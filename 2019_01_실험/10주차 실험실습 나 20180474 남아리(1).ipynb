```python
from uuid import getnode as get_mac
mac = get_mac()
mac_addr=str(mac)
import hashlib
hash_object= hashlib.sha1(mac_addr.encode('utf-8'))
signature = int(hash_object.hexdigest(), 16) % 10**8
print(str(signature))
```

```python
x=np.linspace(0,1,50)
y1=np.cos(4*np.pi*x)
y2=np.cos(4*np.pi*x)*np.exp(-2*x)
plt.subplot(2,1,1)
plt.plot(x,y1,'r-*',lw=1)
plt.ylabel(r'$sin(4\pi x)$')
plt.axis([0,1,-1.5,1.5])
plt.subplot(2,1,2)
plt.plot(x,y2,'b--o',lw=1)
plt.ylabel(r'$e^{-2x}sin(2\pi x)$')
plt.axis([0,1,-1.5,1.5])
plt.show()
```

```python
x=np.linspace(0,1,50)
y1=np.cos(4*np.pi*x)
y2=np.cos(4*np.pi*x)*np.exp(-2*x)

fig, ax = plt.subplots(2,1)
ax[0].plot(x,y1,'r-*',lw=1)
ax[0].set_ylabel(r'$sin(4\pi x)$')
ax[0].axis([0,1,-1.5,1.5])

ax[1].plot(x,y2,'b--o',lw=1)
ax[1].set_ylabel(r'$e^{-2x}sin(2\pi x)$')
ax[1].axis([0,1,-1.5,1.5])

plt.show()
```
