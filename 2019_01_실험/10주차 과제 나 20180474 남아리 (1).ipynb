##  아래 Cell은 각 PC마다 다른 숫자(Hash값)를 생성시키는 Program입니다.
### 실행 ( Run ) 시켜 주십시요...

```python
from uuid import getnode as get_mac
mac = get_mac()
mac_addr=str(mac)
import hashlib
hash_object= hashlib.sha1(mac_addr.encode('utf-8'))
signature = int(hash_object.hexdigest(), 16) % 10**8
print(str(signature))
```

### 과제 2  아래 그림을 그리시요...


```python
t = np.arange(1.,5.,0.25)
plt.plot(t,t,'-kd',label = 't')
plt.plot(t,t*t,'g-o',label = '$t^2$')
plt.axis([1.,5.,0,30])

plt.title('t vs t**2')
plt.ylabel('distance(m)')
plt.xlabel('time(s)')

plt.legend()
plt.show()
```

### 과제 3 아래 그림과 같이 그려보시요... (아래 Hint참조)
-  subplot(2,1,1), subplot(2,1,2) 사용
-  y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
-  y2 = np.cos(2 * np.pi * x2)


```python
x1 = np.arange(0,5,0.1)
x2 = np.arange(0.,2.,0.04)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2,1,1)
plt.plot(x1,y1,'-o')
plt.axis([-0.3,5.3,-0.7,1.1])
plt.title('A table of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(2,1,2)
plt.plot(x2,y2,'-o',markersize=3)
plt.axis([-0.1,2.1,-1.1,1.1])

plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.show()
```
